```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
import ipdb

from process_utils import iou, writer, return_annotate_dict

from sklearn.model_selection import train_test_split

def plot_roc(clf, X, y):
    probs = clf.predict_proba(X)
    preds = probs[:,1]
    fpr, tpr, threshold = metrics.roc_curve(y, preds)
    roc_auc = metrics.auc(fpr, tpr)

    plt.title('Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()

blabels = pd.read_csv('bad_label_losses.csv')
glabels = pd.read_csv('good_label_losses.csv')

blabels['label'] = 0
glabels['label'] = 1

# glabels = glabels[glabels['confidence'] > 0.9]
# blabels = blabels[blabels['confidence'] > 0.9]

train_file = pd.concat((glabels,blabels))

train_file['xy_loss'] = train_file['x_loss'] + train_file['y_loss']
train_file['wh_loss'] = train_file['w_loss'] + train_file['h_loss']

train_file['total_loss'] = train_file['xy_loss'] + train_file['wh_loss']

train_file = train_file.replace([np.inf], np.nan).dropna(subset=['total_loss'], how='all')

X = np.empty(shape=(train_file['total_loss'].shape[0],2))

X[:,0] = train_file['xy_loss']
X[:,1] = train_file['wh_loss']

# X[:,0] = train_file['total_loss']

y = train_file['label']

clf = LogisticRegression()

clf.fit(X,y)

plot_roc(clf, X, y)

labeled_dat = pd.read_csv('labeled_nuro.txt', sep='\t')
gt_dat = pd.read_csv('nuro_gt.txt', sep='\t')

annotation_path = 'updated_train_nuro.txt'
uuid_path = 'updated_uuid_nuro.txt'

merged_dat = pd.merge(labeled_dat, gt_dat, on=['uuid','frame_no', 'camera_no'])

merged_dat['iou'] = merged_dat.apply(lambda row: iou((row['xmin_gt'], row['ymin_gt'], row['xmax_gt'], row['ymax_gt']), 
    (row['x_min'], row['y_min'], row['x_max'], row['y_max'])), axis=1)

train_file = pd.merge(train_file, merged_dat, on=['uuid', 'frame_no']).drop_duplicates()
```

```python
train_file = pd.concat((glabels,blabels))

train_file['xy_loss'] = train_file['x_loss'] + train_file['y_loss']
train_file['wh_loss'] = train_file['w_loss'] + train_file['h_loss']

train_file['total_loss'] = train_file['xy_loss'] + train_file['wh_loss']

train_file = train_file.replace([np.inf], np.nan).dropna(subset=['total_loss'], how='all')

train_file = pd.merge(train_file, merged_dat, on=['uuid', 'frame_no']).drop_duplicates()

train_file.loc[train_file['iou'] <= 0.95, 'label'] = 0

train_file.loc[train_file['iou'] > 0.95, 'label'] = 1

train_file = train_file[train_file['confidence'] > 0.8]

train_file = train_file[train_file['iou'] != 1]

X = np.empty(shape=(train_file['total_loss'].shape[0],2))

X[:,0] = train_file['xy_loss']
X[:,1] = train_file['wh_loss']

y = train_file['label']

xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size = 0.1, random_state = 0)

clf = LogisticRegression(class_weight='balanced')

clf.fit(xTrain,yTrain)

print('Classifier accuracy:', clf.score(xTrain,yTrain))

print('Y mean:', yTrain.mean())

plot_roc(clf, xTest, yTest)

clf.predict(xTest)

from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(100, 50, 25), random_state=1)
clf.fit(xTrain,yTrain)

print('Classifier accuracy:', clf.score(xTest,yTest))

print('Y mean:', yTest.mean())

plot_roc(clf, xTest, yTest)
```
